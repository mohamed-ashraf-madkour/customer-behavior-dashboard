```python
data['Review Rating']=data.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
data.isnull().sum()
```

```python
data.columns=data.columns.str.lower()
```

```python
data.columns=data.columns.str.replace(' ','_')
```

```python
data=data.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
data.columns
```

```python
# create age group
# divide age column to 4 sections
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']

data['age_group'] = pd.qcut(data['age'], q=4, labels=labels)
```

```python
data[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
data['frequency_dayes']=data['frequency_of_purchases'].map(frequency_mapping)
```

```python
data[['frequency_of_purchases','frequency_dayes']]
```

```python
print((data['promo_code_used']==data['discount_applied']).all())
```

```python
data= data.drop('promo_code_used',axis=1)
```
